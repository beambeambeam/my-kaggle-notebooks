# Pinecone Index Listing

This notebook loads `PINECONE_API_KEY` from `.env` and lists the indexes available in your Pinecone account.

```python
from __future__ import annotations

import os
from pathlib import Path

import pandas as pd
from dotenv import load_dotenv
from pinecone import Pinecone

pd.set_option("display.max_colwidth", 200)
pd.set_option("display.width", 220)
```

```python
def load_env_from_parents(start: Path | None = None) -> Path | None:
    start = (start or Path.cwd()).resolve()
    for base in [start, *start.parents]:
        env_path = base / ".env"
        if env_path.exists():
            load_dotenv(env_path, override=False)
            return env_path
    return None


env_path = load_env_from_parents()
api_key = os.getenv("PINECONE_API_KEY")

if not api_key:
    raise RuntimeError(
        "PINECONE_API_KEY was not found. Add it to a .env file or export it in your shell."
    )

print(
    f"Loaded .env from: {env_path}"
    if env_path
    else "No .env file found in parent directories; using current shell environment."
)
```

```python
pc = Pinecone(api_key=api_key)
index_list = pc.list_indexes()

try:
    indexes = index_list.to_dict().get("indexes", [])
except AttributeError:
    indexes = list(index_list)

if not indexes:
    print("No Pinecone indexes found.")
else:
    df = pd.json_normalize(indexes)
    display(df)
    print(f"Found {len(df)} index(es).")
```
