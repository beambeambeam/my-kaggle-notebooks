# Check Documents With More Than One Table

This notebook loads `ocr_table.csv`, counts how many extracted tables belong to each `doc_id`, and shows the documents that have more than one table.

```python
import csv
from collections import Counter, defaultdict
from pathlib import Path
from pprint import pprint

OCR_TABLE_PATH = Path("ocr_table.csv")

rows = []
with OCR_TABLE_PATH.open("r", encoding="utf-8", newline="") as handle:
    reader = csv.DictReader(handle)
    rows = list(reader)

print(f"Loaded {len(rows)} table rows from {OCR_TABLE_PATH}")
rows[:1]
```

```python
table_count_by_doc = Counter(row["id"] for row in rows)
multiple_tables = sorted(
    ((doc_id, count) for doc_id, count in table_count_by_doc.items() if count > 1),
    key=lambda item: (-item[1], item[0]),
)

print(f"Unique doc_id values: {len(table_count_by_doc)}")
print(f"doc_id values with more than one table: {len(multiple_tables)}")
multiple_tables[:20]
```

```python
for doc_id, count in multiple_tables:
    print(f"{doc_id}: {count} tables")
```

```python
rows_by_doc = defaultdict(list)
for row in rows:
    rows_by_doc[row["id"]].append(row)


def show_doc_tables(doc_id: str, preview_chars: int = 500) -> None:
    doc_rows = rows_by_doc.get(doc_id)
    if not doc_rows:
        print(f"No tables found for {doc_id}")
        return

    print(f"{doc_id} has {len(doc_rows)} table(s)\n")
    for index, row in enumerate(doc_rows, start=1):
        print(
            f"Table {index} | section_index={row['section_index']} | table_index={row['table_index']} | tr_count={row['tr_count']}"
        )
        print(row["table_html"][:preview_chars])
        print("\n" + "-" * 80 + "\n")


show_doc_tables("party_list_10_24")
```
